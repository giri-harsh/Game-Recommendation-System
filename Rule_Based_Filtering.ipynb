```python
import pandas as pd
import numpy as np
```

```python
game = pd.read_csv('rawg_games_20k.csv')
game
```

```python
game = game[['id', 'name', 'slug', 'rating', 'platforms', 'genres', 'stores', 'tags']]
```

```python
game.head(1)
```

```python
single = game['tags'].str.contains('singleplayer', case=False, na=False)
multi = game['tags'].str.contains('multiplayer', case=False, na=False)
```

```python
game['play_mode'] = (
    single.map({True: 'SinglePlayer', False: ''}) +
    multi.map({True: 'Multiplayer', False: ''})
)
```

```python
game['play_mode'] = np.select(
    [
        single & multi,
        single & ~multi,
        ~single & multi
    ],
    [
        'Both',
        'Single Player',
        'Multiplayer'
    ],
    default='Unknown'
)
```

```python
game = game.dropna(subset=['genres', 'stores', 'platforms'])
```

```python
game.head(1)
```

```python
def blank_space_remover(obj):
  return obj.str.replace(' ', '', regex=False)
```

```python
game[['platforms','stores','genres']] = game[['platforms','stores','genres']].apply(blank_space_remover)
```

```python
def pipe_remover(obj):
  return obj.str.replace('|', ' ', regex=False)
```

```python
game[['platforms','stores','genres']] = game[['platforms','stores','genres']].apply(pipe_remover)
```

```python
game.head(1)
```

```python
platform_lists = game['platforms'].str.split(' ')
all_platforms = [p.strip() for sublist in platform_lists for p in sublist]
unique_platforms = set(all_platforms)
print(unique_platforms)
```

```python
store_lists = game['stores'].str.split(' ')
all_stores = [p.strip() for sublist in store_lists for p in sublist]
unique_platforms = set(all_stores)
print(unique_platforms)
```

```python
genre_lists = game['genres'].str.split(' ')
all_genres = [p.strip() for sublist in genre_lists for p in sublist]
unique_platforms = set(all_genres)
print(unique_platforms)
```

```python
platform = input("Enter your preferred platform (e.g. PC, PS4, Xbox): ").lower()
genre = input("Enter preferred genre (e.g. Action, RPG, Sports): ").lower()
play_mode = input("Do you prefer singleplayer or multiplayer?: ").lower()
rating = float(input("Minimum rating (e.g. 4.0): "))
store = input("Enter preferred store (e.g. Steam, Origin, Epic Games): ").lower()

filtered = game[
    game['platforms'].str.lower().str.contains(platform) &
    game['genres'].str.lower().str.contains(genre) &
    game['play_mode'].str.lower().str.contains(play_mode) &
    game['stores'].str.lower().str.contains(store) &
    (game['rating'] >= rating)
]
if not filtered.empty:
    print("\nRecommended Games for You:")
    print(filtered[['name', 'genres', 'platforms', 'rating']])
else:
    print("\nNo games found matching your preferences.")
```
