```python
df_copy = df_copy[['id', 'name', 'platforms', 'genres', 'tags']]
```

```python
df_copy = df_copy.dropna(subset=['platforms', 'genres', 'tags'])
```

```python
df_copy
```

```python
df_copy.shape
```

```python
df_copy['platforms_list'] = df_copy['platforms'].str.split('|')
df_copy['genres_list'] = df_copy['genres'].str.split('|')
df_copy['tags_list'] = df_copy['tags'].str.split('|')
```

```python
print("Game:", df_copy.iloc[0]['name'])
print("Platforms:", df_copy.iloc[0]['platforms_list'])
print("Genres:", df_copy.iloc[0]['genres_list'])
print("Tags:", df_copy.iloc[0]['tags_list'][:5])
```

count genre

```python
all_genres = []
for row in df_copy['genres_list']:
    for genre in row:
        all_genres.append(genre)
genre_counts = Counter(all_genres)
```

count platforms

```python
all_platforms = []
for row in df_copy['platforms_list']:
    for platform in row:
        all_platforms.append(platform)
platform_counts = Counter(all_platforms)
```

count tags

```python
all_tags = []
for row in df_copy['tags_list']:
    for tag in row:
        all_tags.append(tag)
tag_counts = Counter(all_tags)
```

```python
top_tags = [tag for tag, count in tag_counts.most_common(75)]
```

```python
print("\nUnique genres:", len(genre_counts))
print("Unique platforms:", len(platform_counts))
print("Unique tags:", len(tag_counts))
print("\nTop 20 tags:")
for i, (tag, count) in enumerate(tag_counts.most_common(20), start=1):
    print(f"{i:2d}. {tag} - {count} times")
```

create binary columns

```python
encoded_df = df_copy[['id', 'name']].copy()
```

```python
# For genres
for genre in genre_counts.keys():
    col_name = f'genre_{genre.replace(" ", "_")}'
    encoded_df[col_name] = df_copy['genres_list'].apply(lambda x: 1 if genre in x else 0)

# For platforms
for platform in platform_counts.keys():
    col_name = f'platform_{platform.replace(" ", "_").replace("/", "_")}'
    encoded_df[col_name] = df_copy['platforms_list'].apply(lambda x: 1 if platform in x else 0)

# For top 75 tags
for tag in top_tags:
    clean_tag = tag.replace(" ", "_").replace("-", "_").replace("/", "_")
    col_name = f'tag_{clean_tag}'
    encoded_df[col_name] = df_copy['tags_list'].apply(lambda x: 1 if tag in x else 0)

print("Binary encoding complete.")
```
